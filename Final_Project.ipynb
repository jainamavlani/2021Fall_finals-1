```python
w = 170
h = 50
d = 70
plt.rc('font', size=50)          # controls default text sizes
plt.rc('axes', titlesize=50)     # fontsize of the axes title
plt.rc('axes', labelsize=50)    # fontsize of the x and y labels and title
plt.rc('xtick', labelsize=50)    # fontsize of the tick labels
plt.rc('ytick', labelsize=50)
axis_font = {'fontname':'Arial', 'size':'90'}
plt.figure(figsize=(w, h), dpi=d)
plt.plot(gdp_USA['variable'],gdp_USA['value'])
plt.xlabel("Year", **axis_font)
plt.ylabel("GDP (xle13)", **axis_font)
plt.title("GDP of the USA over the years(Graph 1A)",**axis_font)
#plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
#plt.rc('ytick', labelsize=SMALL_SIZE)
```

```python
total_loan = pd.read_csv('BUSLOANS.csv')
total_loan["DATE"]= pd.to_datetime(total_loan["DATE"])
total_loan = total_loan.groupby(total_loan['DATE'].dt.year)['BUSLOANS'].agg(['sum'])
total_loan['Loan_BUSD'] = total_loan['sum']
total_loan.drop('sum', axis=1, inplace=True)
total_loan.head()


```

```python
w = 170
h = 50
d = 70
plt.rc('font', size=50)          # controls default text sizes
plt.rc('axes', titlesize=50)     # fontsize of the axes title
plt.rc('axes', labelsize=50)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=50)    # fontsize of the tick labels
plt.rc('ytick', labelsize=50)
axis_font = {'fontname':'Arial', 'size':'90'}
plt.figure(figsize=(w, h), dpi=d)
plt.plot(total_loan['Loan_BUSD'])
plt.xlabel("Year", **axis_font)
plt.ylabel("Total Loan in Billion USD", **axis_font)
plt.title("Total loan taken in the USA over the years(Graph 1B)",**axis_font)
```

As shown above in the Graph 1A, the GDP has increased over the years post the WWII noting that low interest rates were offered in start of the period post the WWII.
As shown above in the Graph 1B, the total loan that was taken over the years post WWII has increased. However there is a sloght downward trend scene arounf 2009 due to the great recession and around 2020 due to COVID.
Hence we can say that as the habbit of taking a loan started getting formed, the GDP increased with increase in total loan that was taken in the USA.
Hence we can say that the null hypotheis is satisfied.

```python
income_expenditure = pd.merge(inc, personal_expenditure,how='inner', on='Year')

income_expenditure['income_expenditure_diff'] = income_expenditure['Current Dollars'] - income_expenditure['sum']
```

```python
w = 170
h = 50
d = 70
plt.rc('font', size=50)          # controls default text sizes
plt.rc('axes', titlesize=50)     # fontsize of the axes title
plt.rc('axes', labelsize=50)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=50)    # fontsize of the tick labels
plt.rc('ytick', labelsize=50)
axis_font = {'fontname':'Arial', 'size':'90'}
plt.figure(figsize=(w, h), dpi=d)
plt.plot(income_expenditure['Year'],income_expenditure['income_expenditure_diff'])
plt.xlabel("Year", **axis_font)
plt.ylabel("income - personal_expenditure", **axis_font)
plt.title("Graph showing the decreasing trend in the difference between income and personal expenditure over the years (2c)",**axis_font)
```

As shown in the graph above, imposing less interest in the initial years led to the formation of habit of spending more than you earn. This led to people spending more than they earned. This formation of habit of spending so much, led to the building of the GDP in countries like the USA.
Since the Difference between the income and personal expenditure has gradually decreased over the years after the passing of the GI bill and introduction of the credit system,the hypothesis 2 holds true.



This also prooves the notions which says: "You should live a lifestyle that most of the other Amercans live. This can be fueled by going into debt."

By following the notion mentioned above, we can say that if one would live a lifestyle that most other Americans lived, even a person who lies in the 50th percentile of income would try to live a lifestyle that a person in the 90th percentile lived. This would eventually lead to poor getting poorer, since they would have to spend a larger proportion of their income to keep up with lifestyle and spending of the people in the higher income bracket.
